```python
import selenium
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By

import selenium
from selenium import webdriver
import time
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# Let's first connect to the driver
driver = webdriver.Chrome("chromedriver.exe")
#opening the cars24 website
url = 'https://www.cars24.com/buy-used-car?sort=P&storeCityId=2&pinId=110001'
driver.get(url)
```

```python
i=0
while(i<100):
    driver.execute_script("window.scrollBy(0,500)") # scroll down to get more comments
    i+=1
while(i<5002):
    driver.execute_script("window.scrollBy(0,5000)") # scroll down to get more comments
    i+=1
```

```python
# Scraping name of car
name = []
title_tags = driver.find_elements(By.XPATH,'//h3[@class="_3FpCg"]')
for i in title_tags:
    title = i.text
    name.append(title)
len(name)
```

```python
# Scraping the year
year = []
title_tags = driver.find_elements(By.XPATH,'//h3[@class="_3FpCg"]')
for i in title_tags:
    title = i.text
    year.append(title.split(" ")[0])
len(year)
```

```python
# Scraping price of the car
price = []
title_tags = driver.find_elements(By.XPATH,'//div[@class="_7udZZ"]')
for i in title_tags:
    title = i.text
    price.append(title)
len(price)
```

```python
# scraping the kms
k = []
title_tags = driver.find_elements(By.XPATH,'//ul[@class="bVR0c"]/li[1]')
for i in title_tags:
    title = i.text.replace('km','')
    k.append(title)
len(k)
```

```python
# scraping the fuel type
fuel = []
title_tags = driver.find_elements(By.XPATH,'//ul[@class="bVR0c"]/li[3]')
for i in title_tags:
    title = i.text
    fuel.append(title)
len(fuel)
```

```python
transmission = []
title_tags = driver.find_elements(By.XPATH,'//p[@class="cvakB"]/span')
for i in title_tags:
    title = i.text
    transmission.append(title)
len(transmission)
```

```python
owner = []
title_tags = driver.find_elements(By.XPATH,"//ul[@class='bVR0c']/li[2]")
for i in title_tags:
    title = i.text
    owner.append(title)
len(owner)
```

```python
#Creating a dataframe and checking the data extracted
cars24_delhi = pd.DataFrame({'name':name,'Year':year,'selling_price':price,'km_driven':k,'fuel':fuel,'transmission':transmission,'owner':owner})
```

```python
cars24_delhi['State']='Delhi'
```
